```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier as KNN
from time import time
from sklearn.linear_model import LogisticRegression
import pickle
import math
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("Test1655_240518.CSV") 
df.describe()
```

```python
x_values=df.iloc[:,0].values
y_values=df.iloc[:,1].values
```

## Training model

```python
j=0
i=0
labels=[];
feature1=[];
feature2=[];
feature3=[];
feature4=[];
count=0
end=0
start=5930;
while(start>=5930 and end<17930):
    #FAN SPEED 1 and FAN SPEED 2

    labels.append(1)
    end=start+100; 
    feature1.append(np.mean(x_values[start:end]))
    feature2.append(np.std(x_values[start:end])) 
    feature3.append(np.mean(y_values[start:end]))
    feature4.append(np.std(y_values[start:end])) 
    start=end
 


feature1=(abs(np.diff(feature1))).tolist()
feature2=(abs(np.diff(feature2))).tolist()
feature3=(abs(np.diff(feature3))).tolist()
feature4=(abs(np.diff(feature4))).tolist()
labels=labels[:-1]


start=20930
end=0



while(start>=20930 and end<23930):

    #POWDER
    end=start+100
    feature1.append(np.mean(x_values[start:end]))
    feature2.append(np.std(x_values[start:end]))
    feature3.append(np.mean(y_values[start:end]))
    feature4.append(np.std(y_values[start:end]))
    start=end
 
    labels.append(2)


    
feature1=(abs(np.diff(feature1))).tolist()
feature2=(abs(np.diff(feature2))).tolist()
feature3=(abs(np.diff(feature3))).tolist()
feature4=(abs(np.diff(feature4))).tolist()
labels=labels[:-1]

m=26930
n=0


while(m>=26930 and n<29930):
    ##Spray

    n=m+100
    feature1.append(np.mean(x_values[m:n]))
    feature2.append(np.std(x_values[m:n]))
    feature3.append(np.mean(y_values[m:n]))
    feature4.append(np.std(y_values[m:n])) 
    labels.append(3)
    m=n

feature1=(abs(np.diff(feature1)))
feature2=(abs(np.diff(feature2)))
feature3=(abs(np.diff(feature3)))
feature4=(abs(np.diff(feature4)))
labels=labels[:-1]

mag_diff1=(np.sqrt(feature1**2+feature3**2)).reshape(feature1.shape[0],1);
dir_diff1=(np.arctan(feature3/feature1)).reshape(feature1.shape[0],1);

mag_diff2=(np.sqrt(feature2**2+feature4**2)).reshape(feature1.shape[0],1);
dir_diff2=(np.arctan(feature4/feature2)).reshape(feature1.shape[0],1);

# Training features

training=np.concatenate((mag_diff1,dir_diff1,mag_diff2,dir_diff2), axis=1)
testing=np.asarray(labels)
print training.shape
print testing.shape


#TRAINING THE MODEL
clf = RandomForestClassifier(random_state=1)
clf.fit(training,testing)


with open('Random_Forest_MODE.pickle','wb') as f:
    pickle.dump(clf,f)


print "MODEL SAVED"
```
